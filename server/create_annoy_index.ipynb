```python
from annoy import AnnoyIndex
import numpy as np
import pandas as pd
```

```python
flows = np.load('flows.npy', allow_pickle=True)
```

```python
t = AnnoyIndex(500, 'euclidean')
```

```python
for i,row in enumerate(flows):
    flows_df = pd.DataFrame(row)
    flows_df = flows_df[[1,2]]
    flows_arr = flows_df.to_numpy().reshape(-1)
    flows_arr = np.pad(flows_arr, (0,500-len(flows_arr)))
    t.add_item(i,flows_arr)
```

```python
t.build(50)
```

```python
t.save('flows.ann')
```
